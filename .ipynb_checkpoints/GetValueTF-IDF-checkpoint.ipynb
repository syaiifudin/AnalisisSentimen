```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
data = pd.read_csv('Processing.csv')
data
```

```python
count = CountVectorizer()
titles = pd.read_csv('Processing.csv', sep=',' ,usecols=['text'], squeeze=True)
docs = titles.values
bag = count.fit_transform(docs)
print(docs)
```

```python
type(docs)
docs.shape
```

```python
print(count.vocabulary_)
```

```python
print(count.get_feature_names())
a=count.get_feature_names()
```

```python
print(bag.toarray())
b = bag.toarray()
```

```python
dfa = pd.DataFrame(data=a)
num_rows = -1
for row in open("Processing.csv"):
    num_rows +=1
    
dfb = pd.DataFrame(data=b, index=range(0, num_rows), columns=[a])
dfb
```

```python
dfb.to_excel("PembobotanTF.xlsx")  
```

```python
tfidf = TfidfTransformer(use_idf=True,
                         norm ='l2',
                         smooth_idf=True)

tf = tfidf.fit_transform(count.fit_transform(docs)).toarray()

num_rows = -1
for row in open("Processing.csv"):
    num_rows +=1
    
dfbtf = pd.DataFrame(data=tf, index=range(0,num_rows), columns=[a])
dfbtf
```

```python
dfbtf.to_excel("PembobotanTF-IDF.xlsx")  
```
